```python
import math
import pandas as pd

with open("01_corpus_turismo.txt", "r") as f:
    corpus = f.readlines()

corpus = [linea.strip() for linea in corpus]

diccionario = []
for doc in corpus:
    for palabra in doc.split():
        if palabra not in diccionario:
            diccionario.append(palabra)

matriz_td = []
valores_df = []
valores_idf = []

N = len(corpus)

for termino in diccionario:
    array = []
    df_termino = 0
    for doc in corpus:
        count = doc.split().count(termino)
        array.append(count)
        if count > 0:
            df_termino += 1
    matriz_td.append(array)
    valores_df.append(df_termino)

    valores_idf.append(math.log(N / df_termino) if df_termino > 0 else 0)

num_docs = [f"Documento{i+1}" for i in range(N)]
data_td = pd.DataFrame(matriz_td, index=diccionario, columns=num_docs)

data_idf = pd.DataFrame({
    'Término': diccionario,
    'DF': valores_df,
    'IDF': valores_idf
})
data_idf.set_index('Término', inplace=True)

print("Matriz Termino - Documento:")
display(data_td)

print("\nIDF por Término:")
display(data_idf)
```
